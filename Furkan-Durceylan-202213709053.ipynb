**Soru 3: Needleman-Wunsch Algoritması (20 puan)**

Amaç: Needleman-Wunsch algoritması, iki biyolojik diziyi (örneğin DNA veya protein dizilerini) en iyi şekilde global olarak hizalamak için kullanılan bir dinamik programlama algoritmasıdır. Bu algoritma, dizi eşleşmelerinde benzerliklerin maksimize edilmesi için özellikle kullanışlıdır.

Adımlar:

1. Matris Hazırlama: İlk olarak, dizi uzunluklarına bağlı olarak bir matris oluşturulur. İlk satır ve sütun boşluk cezaları ile doldurulur.
2. Skorlandırma: İki dizi arasındaki her harf eşleşmesi için, uygun skorlar hesaplanır:
  Eşleşme: Harfler aynı ise belirlenen bir eşleşme skoru eklenir.

  Boşluk (Gap) Cezası: Dizi boyunca boşluk olduğunda uygulanır.

  Uyumsuzluk: Harfler farklı ise uyumsuzluk skoru uygulanır.
3. Geri İzleme: Matristeki en büyük skordan başlanarak en kısa yoldan köşeye (başlangıç noktasına) gidilir. Bu yolla dizilerdeki uyum ve boşluklar belirlenir.

Örnek

Dizi 1: GATTACA

Dizi 2: GCATGCU

Eşleşme Skoru: +1

Uyumsuzluk Skoru: -1

Boşluk Cezası: -1

1. Matris oluşturulup boşluklar ve başlangıç değerleri yerleştirilir.
2. Skorlar, yukarıda anlatılan kurallara göre hesaplanır ve matris doldurulur.
3. Geri izleme ile en iyi hizalama belirlenir.

**Bu ve benzeri bilgileri araştırınız ve aşağıdaki soruyu algoritmaya uyarlayınız.**

Verilen iki dizi için (Dizi1: AGACTG, Dizi2: AGCTG) Needleman-Wunsch algoritmasını kullanarak global hizalamayı sağlayan bir Python fonksiyonu yazın. Aşağıdaki parametreleri kullanın:

Eşleşme Skoru: +2

Uyumsuzluk Skoru: -1

Boşluk Cezası: -2

> yardımcı kaynak : https://en.wikipedia.org/wiki/Needleman–Wunsch_algorithm



````python
def needleman_wunsch(seq1, seq2, match=2, mismatch=-1, gap=-2):

    n = len(seq1)
    m = len(seq2)

    score = [[0] * (m + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        score[i][0] = score[i - 1][0] + gap
    for j in range(1, m + 1):
        score[0][j] = score[0][j - 1] + gap


    for i in range(1, n + 1):
        for j in range(1, m + 1):

            if seq1[i - 1] == seq2[j - 1]:
                diag = score[i - 1][j - 1] + match
            else:
                diag = score[i - 1][j - 1] + mismatch

            up = score[i - 1][j] + gap
            left = score[i][j - 1] + gap

            score[i][j] = max(diag, up, left)


    i, j = n, m
    align1 = []
    align2 = []

    while i > 0 or j > 0:

        if i > 0 and j > 0:
            if seq1[i - 1] == seq2[j - 1]:
                diag_score = score[i - 1][j - 1] + match
            else:
                diag_score = score[i - 1][j - 1] + mismatch

            if score[i][j] == diag_score:

                align1.append(seq1[i - 1])
                align2.append(seq2[j - 1])
                i -= 1
                j -= 1
                continue

        if i > 0 and score[i][j] == score[i - 1][j] + gap:
            align1.append(seq1[i - 1])
            align2.append('-')
            i -= 1
        else:

            align1.append('-')
            align2.append(seq2[j - 1])
            j -= 1


    align1 = ''.join(reversed(align1))
    align2 = ''.join(reversed(align2))


    final_score = score[n][m]

    return score, align1, align2, final_score



# Örnek kullanım
seq1 = "AGACTG"
seq2 = "AGCTG"

score_matrix, aln1, aln2, total = needleman_wunsch(seq1, seq2,
                                                   match=2,
                                                   mismatch=-1,
                                                   gap=-2)

print("Hizalama 1:", aln1)
print("Hizalama 2:", aln2)
print("Toplam skor:", total)

````

**Soru 4: Smith-Waterman Algoritması (20 puan)**

Amaç: Smith-Waterman algoritması, iki biyolojik dizi arasında en iyi yerel hizalamayı bulmak için kullanılır. Bu algoritma, dizinin tamamının değil yalnızca benzer olan kısımlarının hizalanmasını sağlar.

Adımlar:
1. Matris Hazırlama: Dizi uzunluklarına göre bir matris oluşturulur. İlk satır ve sütun sıfırla doldurulur.
2. Skorlandırma: Her hücre, komşu hücrelerden maksimum skor alınarak doldurulur:

  Yukarıdaki hücreden gelen skor (boşluk cezası eklenir).
  
  Sol hücreden gelen skor (boşluk cezası eklenir).

  Çapraz hücreden gelen skor (eşleşme/uyumsuzluk skoru eklenir).

  Negatif skorlara izin verilmez, en düşük değer 0 olur.
3. Geri İzleme: En yüksek skorlu hücreden başlayarak 0 değerine ulaşana kadar geri izleme yapılır.

Örnek

Dizi 1: GATTACA

Dizi 2: GCATGCU

Eşleşme Skoru: +1
Uyumsuzluk Skoru: -1
Boşluk Cezası: -1

Matris doldurulur ve negatif skor yerine 0 kullanılır.

En yüksek skordan 0 değerine kadar geri izlenir.

**Bu ve benzeri bilgileri araştırınız ve aşağıdaki soruyu algoritmaya uyarlayınız.**

Verilen iki dizi için (Dizi1: AGACTG, Dizi2: CGTGA) Smith-Waterman algoritmasını kullanarak yerel hizalamayı sağlayan bir Python fonksiyonu yazın. Aşağıdaki parametreleri kullanın:

Eşleşme Skoru: +3

Uyumsuzluk Skoru: -3

Boşluk Cezası: -2



> yardımcı kaynak : https://en.wikipedia.org/wiki/Smith–Waterman_algorithm





````python
def smith_waterman(seq1, seq2, match=3, mismatch=-3, gap=-2):

    n = len(seq1)
    m = len(seq2)


    score = [[0] * (m + 1) for _ in range(n + 1)]
    pointer = [[""] * (m + 1) for _ in range(n + 1)]  # "diag", "up", "left", "" (stop)

    max_score = 0
    max_pos = (0, 0)


    for i in range(1, n + 1):
        for j in range(1, m + 1):

            if seq1[i - 1] == seq2[j - 1]:
                diag_score = score[i - 1][j - 1] + match
            else:
                diag_score = score[i - 1][j - 1] + mismatch

            up_score = score[i - 1][j] + gap
            left_score = score[i][j - 1] + gap


            best = max(0, diag_score, up_score, left_score)
            score[i][j] = best


            if best == 0:
                pointer[i][j] = ""
            elif best == diag_score:
                pointer[i][j] = "diag"
            elif best == up_score:
                pointer[i][j] = "up"
            elif best == left_score:
                pointer[i][j] = "left"


            if best > max_score:
                max_score = best
                max_pos = (i, j)


    i, j = max_pos
    align1 = []
    align2 = []

    while i > 0 and j > 0 and score[i][j] != 0:
        direction = pointer[i][j]

        if direction == "diag":
            align1.append(seq1[i - 1])
            align2.append(seq2[j - 1])
            i -= 1
            j -= 1
        elif direction == "up":
            align1.append(seq1[i - 1])
            align2.append('-')
            i -= 1
        elif direction == "left":
            align1.append('-')
            align2.append(seq2[j - 1])
            j -= 1
        else:
            break


    align1 = ''.join(reversed(align1))
    align2 = ''.join(reversed(align2))

    return score, align1, align2, max_score



# Örnek kullanım
seq1 = "AGACTG"
seq2 = "CGTGA"

score_matrix, aln1, aln2, best_score = smith_waterman(
    seq1, seq2,
    match=3,
    mismatch=-3,
    gap=-2
)

print("Yerel hizalama 1:", aln1)
print("Yerel hizalama 2:", aln2)
print("En yüksek skor (yerel hizalama skoru):", best_score)























# Örnek kullanım
seq1 = "AGACTG"
seq2 = "CGTGA"






````

**Soru 5: Motif Arama ve Frekans Analizi (20 puan)**

Açıklama: Biyoinformatikte, belirli bir DNA veya RNA dizisinde biyolojik işlevi olan belirli motifleri aramak oldukça önemlidir. Bu soruda, bir DNA dizisinde verilen bir motifin tüm konumlarını ve dizideki tekrar sayısını bulmanız gerekecektir.

Görev:
Aşağıdaki adımları izleyerek bir Python programı yazın:

1. dna_sequence adında bir DNA dizisini ve motif adında bir motif dizisini kullanıcıdan alarak iki değişken oluşturun.
2. motif_arama adında bir fonksiyon tanımlayın ve bu fonksiyon:
DNA dizisi içerisinde motifin tüm başlangıç indekslerini bulsun.
Motifin dizide kaç kez tekrar ettiğini hesaplasın.
3. Fonksiyon, bulunan tüm indekslerin bir listesini ve motif tekrar sayısını döndürmelidir.

**Örnek DNA ve Motif Dizisi:**

dna_sequence = "ATGCTTCAGATGCTTGGATGCGT"

motif = "ATG"

Beklenen Çıktı olarak fonksiyon, aşağıdaki çıktıyı vermelidir:

Motifin bulunduğu başlangıç indeksleri: [0, 9, 17]

Motifin tekrar sayısı: 3






````python
def motif_arama(DnaSeq, Motif):
    indeksler = []
    sayac = 0
    mot_len = len(Motif)

    for i in range(len(DnaSeq) - mot_len + 1):
        if DnaSeq[i:i + mot_len] == Motif:
            indeksler.append(i)
            sayac += 1

    return indeksler, sayac

# Kullanıcı girdisi
dna_sequence = input("DNA dizisini girin: ")
motif = input("Aranacak motifi girin: ")

# Motif arama işlemini gerçekleştirin

indeksler, tekrar_sayisi = motif_arama(dna_sequence, motif)




# Sonuçları ekrana yazdırın

print("Motifin bulunduğu başlangıç indeksleri:", indeksler)
print("Motifin tekrar sayısı:", tekrar_sayisi)



````
